```python
import cv2
from google.colab.patches import cv2_imshow
import math

```

```python
img = cv2.imread('example3.jpg')
cv2_imshow(img)
```

```python
def gamma_correction(y, img):
  corrected_img = img
  print("y - ", y )
  for i in range(len(corrected_img)):
    for k in range(len(corrected_img[i])):
      for rgb in range(len(corrected_img[i][k])):
        corrected_img[i][k][rgb] = math.pow(float(corrected_img[i][k][rgb]) / 255.0, 1/y) * 255.0
  return corrected_img
```

```python

img1 = gamma_correction(0.9, img)
cv2_imshow(img1)
```
